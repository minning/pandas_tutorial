```python
import pandas as pd
import numpy as np
```

```python
data_path = 'datas/ml-1m/ratings.dat'
```

```python
df = pd.read_csv(data_path,
                sep='::',
                engine='python',
                names=['UserId','MovieId','Rating','TimeStap'])
```

```python
df.shape
```

```python
df.head()
```

```python
def get_norm_rate(df):
    max_rate = df['Rating'].max()
    min_rate = df['Rating'].min()
    
    df['norm_rate'] = df['Rating'].apply(lambda x:(x-min_rate)*1.0/(max_rate-min_rate))
    return df
df = df.groupby('UserId').apply(get_norm_rate)
```
