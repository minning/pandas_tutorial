1、map只用于Series，事先值到值的映射
2、apply用于Series事先值到值的映射，用于DataFrame实现某个轴的Series的处理
3、applymap只能用于DataFrame，用于处理DataFrame中每一个元素

```python
import numpy as np
import pandas as pd
```

```python
data_path = 'datas/stock/互联网公司股票历史数据.csv'
```

```python
df = pd.read_csv(data_path)
```

```python
df
```

```python
df.dropna(axis='index',inplace=True)
```

```python
df
```

```python
df['公司'].unique()
```

```python
company_map = {
    '美团' : 'meituan',
    '腾讯' : 'tencent',
    '百度' : 'baidu'
}
```

```python
df['company_eng'] = df['公司'].map(company_map)
```

```python
df
```

```python
df['company_eng2'] = df['公司'].apply(lambda x:company_map[x])
```

```python
df
```

```python
df['company_eng3'] = df.apply(lambda x : company_map[x['公司']], axis=1)
```

```python
df_sub = df[['收盘','开盘','高','低']]
```

```python
df_sub
```

```python
df_sub.applymap(lambda x:int(x))
```

```python
df.loc[:,['收盘','开盘','高','低']] = df_sub.applymap(lambda x:int(x))
```
