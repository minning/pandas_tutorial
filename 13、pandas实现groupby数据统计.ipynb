```python
data_path = 'datas/beijing_tianqi/beijing_tianqi_2018.csv'
```

```python
df = pd.read_csv(data_path)
```

```python
df
```

```python
df['bwendu'] = df['bwendu'].str.replace('C', '').astype(np.int32)
```

```python
df['ywendu'] = df['ywendu'].str.replace('C', '').astype(np.int32)
```

```python
df
```

```python
df['month'] = df['ymd'].str.replace('-','').str[:6]
```

```python
df
```

```python
df['day'] = df['ymd'].str.replace('-', '').astype(np.int32)
```

```python
df
```

```python
# 每天最高温度
data = df.groupby('day',as_index=False)['bwendu'].max()
```

```python
data = data.rename(columns={'bwendu':'wendu_max'})
```

```python
data
```

```python
df = pd.merge(df,data,on='day')
```
