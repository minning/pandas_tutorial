```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('datas/beijing_tianqi/beijing_tianqi_2018.csv')
```

```python
df
```

```python
df['bwendu'] = df['bwendu'].map(lambda x:x.replace('C', '')).astype(np.int16)
```

```python
df['ywendu'] = df['ywendu'].map(lambda x:x.replace('C', '')).astype(np.int16)
```

```python
df
```

```python
def wencha(x):
    return x['bwendu'] - x['ywendu']
df['wencha'] = df.apply(wencha, axis=1)
```

```python
df
```

```python
def wencha_and_avg(x):
    return x['bwendu'] - x['ywendu'], (x['bwendu'] + x['ywendu'])/2
df[['wencha2', 'wendu_avg']] = df.apply(wencha_and_avg, axis=1, result_type='expand')
```
